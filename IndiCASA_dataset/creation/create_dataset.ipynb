```python
import pickle

with open("datasets/caste_data.pkl", "rb") as f:
    caste = pickle.load(f)
with open("datasets/religion_data.pkl", "rb") as f:
    religion = pickle.load(f)
with open("datasets/disability_data.pkl", "rb") as f: 
    disability = pickle.load(f)
with open("datasets/gender_data.pkl", "rb") as f:
    gender = pickle.load(f)
with open("datasets/socioeconomic_data.pkl", "rb") as f:
    socioeconomic = pickle.load(f)
```

```python
caste[1]
```

```python
from datasets import Dataset, DatasetDict

# 1. define your bias‐types
biases = {
    "caste" : caste, 
    "religion" : religion, 
    "disability" : disability, 
    "gender" : gender,
    "socioeconomic" : socioeconomic
}

# 2. for each bias load its stereo & anti lists and build examples
all_splits = {}
for bias in biases.keys():
    stereo_lists, anti_lists = biases[bias]  
    examples = [
        {
            "context_id": i,
            "stereotypes": stereo,
            "anti_stereotypes": anti,
        }
        for i, (stereo, anti) in enumerate(zip(stereo_lists, anti_lists))
    ]
    # wrap each as its own Dataset
    all_splits[bias] = Dataset.from_list(examples)

# 3. build the DatasetDict
ds_dict = DatasetDict(all_splits)
```
